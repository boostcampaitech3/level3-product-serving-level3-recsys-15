```python
# 교보문고 데이터 크롤링
s=Service('/opt/ml/chromedriver')
driver = webdriver.Chrome(options=options, service=s)
driver.implicitly_wait(20)
driver.get("http://www.kyobobook.co.kr/index.laf")

for book in range(len(books)):
    keywords = ''
    lines = ''
    topics = ''
    category = ''
    try:
        driver.find_element_by_xpath('//*[@id="searchKeyword"]').send_keys(books[book])
        driver.find_element_by_xpath('//*[@id="searchTop"]/div/input').click()
        driver.find_element_by_xpath('//*[@id="search_list"]/tr/td[2]/div[2]/a').click()
        keywords = ''
        keyword = driver.find_element_by_xpath('//*[@id="container"]/div[2]/form/div[2]/div[3]/div[2]')
        for i in keyword.find_elements_by_xpath('.//*'):
            keywords += i.text
            keywords += ' '

        topics = driver.find_element_by_xpath('//*[@id="container"]/div[5]/div[1]/div[3]/div[2]').text
        driver.find_element_by_xpath('//*[@id="container"]/div[5]/div[1]/div[3]/div[10]/div[1]/a').click()
        lines = driver.find_element_by_xpath('//*[@id="container"]/div[5]/div[1]/div[3]/div[10]/div[2]').text
        category = driver.find_element_by_xpath('//*[@id="container"]/div[5]/div[1]/div[3]/ul/li').text
    except:
        print(book, 'error')
        keywords = ''
        lines = ''
        topics = ''
        category = ''
    books_info[books[book]].append(keywords)
    books_info[books[book]].append(topics)
    books_info[books[book]].append(lines)
    books_info[books[book]].append(category)

    if book % 10 == 0:
        print(book, "is completed")
```

```python
name = []
isbn = []
writer = []
genre = []
images = []
raw = []

s=Service('/opt/ml/chromedriver')
driver = webdriver.Chrome(options=options, service=s)
driver.implicitly_wait(20)
driver.get("https://www.yes24.com/Templates/FTLogin.aspx")
driver.find_element_by_xpath('//*[@id="SMemberID"]').send_keys('')
driver.find_element_by_xpath('//*[@id="SMemberPassword"]').send_keys('')
driver.find_element_by_xpath('//*[@id="btnLogin"]/span').click()

for page in range(1, 81):
    driver.get("http://www.yes24.com/24/Category/Display/001001046004?PageNumber={}".format(page))

    # driver.find_element_by_xpath('//*[@id="bigbanner_domestic_area"]/h3/a').click()
    # driver.find_element_by_xpath('//*[@id="mCateLi"]/li[8]/a').click()
    # driver.find_element_by_xpath('//*[@id="cateSubListWrap"]/dl[8]/dt/a').send_keys(Keys.ENTER)
    for i in range(1, 21):
        try:
            driver.find_element_by_xpath('//*[@id="category_layout"]/ul/li[{}]/div/p/span/span/a'.format(i)).send_keys(Keys.ENTER)
        except WDE:
            print('error')
        try:
            n = driver.find_elements_by_class_name('gd_name')[0].text
            ib = driver.find_element_by_xpath('//*[@id="infoset_specific"]/div[2]/div/table/tbody/tr[3]/td').text
            w = driver.find_element_by_class_name('gd_auth').text
            g = driver.find_element_by_xpath('//*[@id="infoset_goodsCate"]/div[2]/dl/dd/ul').text
            im = driver.find_element_by_xpath('//*[@id="yDetailTopWrap"]/div[1]/div/span/em/img').get_attribute('src')
            r = driver.find_element_by_xpath('/html/head/meta[10]').get_attribute('content')
        except NoSuchElementException:
            print("NoSuchElementException")
            driver.execute_script("window.history.go(-1)")
            continue
        name.append(n)
        isbn.append(ib)
        writer.append(w)
        genre.append(g)
        images.append(im)
        raw.append(r)
        driver.execute_script("window.history.go(-1)")
        print(name[-1])
    time.sleep(1)
driver.quit()

data = pd.DataFrame()
data["name"] = name
data['isbn'] = isbn
data['writer'] = writer
data['genre'] = genre
data['images'] = images
data['raw'] = raw
data.to_csv("china.csv", index=False)
```
