```python
from selenium import webdriver
from selenium.webdriver.chrome import service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import WebDriverException as WDE
from selenium.common.exceptions import NoSuchElementException, InvalidSessionIdException, UnexpectedAlertPresentException


import pandas as pd
import time
```

```python
options = Options()
options.add_argument("--no-sandbox") #bypass OS security model
options.add_argument("--headless")
options.add_argument("--disable-dev-shm-usage") #overcome limited resource problems
```

# 교보문고 크롤링

```python
# 교보문고 크롤링
nara = ['japan', 'china', 'france','german','russia','spanish','north_eu','other']

for n in nara:
    data = pd.read_csv('{}.csv'.format(n)) # yes24 data 저장한 파일 불러오기
    isbns = data['isbn'] # yes24 data isbn 파일 불러오기
    keywordpick = []
    keywords = []
    isbn_t = []
    text = []

    s=Service('/opt/ml/chromedriver')
    driver = webdriver.Chrome(options=options, service=s)
    driver.implicitly_wait(5)
    # 로그인
    driver.get('http://www.kyobobook.co.kr/login/login.laf?Kc=GNHHNOlogin&orderClick=c03&retURL=http%3A//www.kyobobook.co.kr/index.laf')
    driver.find_element_by_xpath('//*[@id="memid"]').send_keys('') # id 
    driver.find_element_by_xpath('//*[@id="pw"]').send_keys('') # 비밀번호
    driver.find_element_by_xpath('//*[@id="login_zone1"]/p[2]/input').click()



    for count, isbn in enumerate(isbns):
        if count % 100 == 99:
            driver.quit()
            s=Service('/opt/ml/chromedriver')
            driver = webdriver.Chrome(options=options, service=s)
            driver.implicitly_wait(5)
            # 로그인
            driver.get('http://www.kyobobook.co.kr/login/login.laf?Kc=GNHHNOlogin&orderClick=c03&retURL=http%3A//www.kyobobook.co.kr/index.laf')
            driver.find_element_by_xpath('//*[@id="memid"]').send_keys('') # id
            driver.find_element_by_xpath('//*[@id="pw"]').send_keys('') # 비밀번호
            driver.find_element_by_xpath('//*[@id="login_zone1"]/p[2]/input').click()

        driver.get(url = "http://www.kyobobook.co.kr/product/detailViewKor.laf?mallGb=KOR&ejkGb=KOR&barcode=" + str(isbn) + "&orderClick=JAI")
        # 키워드 픽
        try:
            keyword = driver.find_element_by_class_name('box_detail_keywordpick')
            k = ''
            for i in keyword.find_element_by_class_name('book_keyword').find_elements_by_xpath(".//*"):
                k += i.text
                k += ' '
            
        except NoSuchElementException:
            print("No keywordpick")
            continue
        except UnexpectedAlertPresentException:
            print("UnexpectedAlertPresentException")
            continue

        # 키워드
        try:
            
            key = driver.find_element_by_name('keywords').get_attribute('content')
            context = driver.find_element_by_xpath('//*[@id="container"]/div[5]/div[1]/div[3]').get_attribute('textContent')
        except NoSuchElementException:
            print("NoSuchElementException")
            continue
        isbn_t.append(isbn)
        keywords.append(key)
        text.append(context)
        keywordpick.append(k)
        print(n, count)

    driver.quit()

    data = pd.DataFrame()
    data['isbn'] = isbn_t
    data['keywordpick'] = keywordpick
    data['keywords'] = keywords
    data['text'] = text
    data.to_csv("{}_kyobo.csv".format(n), index=False)

```

# 교보문고 데이터 정제

```python
import re

# 북믈리에
b = re.compile("북소믈리에 한마디!")
# 목차
m = re.compile("\n\t목차\n\t")
# 책속으로
c = re.compile('\n\t책 속으로\n\t')
# 출판사 서평
p = re.compile('\n\t출판사 서평\n\t')

# 닫기
d = re.compile('\t\t닫기\n\t')

# 더보기
mo = re.compile('\t\t더보기\n\t')

# 종료 (북소믈리에만)
e = re.compile('\n\t\t\t\t\n\t\t\t\n\t\t\n\t\t\n\t\t\n\t\t\t\n\t\t\t\t')
```

```python
nara = ['japan', 'china', 'france','german','russia','spanish','north_eu','other']

for n in nara:
    data = pd.read_csv('{}_kyobo.csv'.format(n))

    isbn = data['isbn']
    keywords = data['keywords']
    text= data['text']

    # 키워드 스플릿
    new_keywords = []
    for i, k in enumerate(keywords):
        split_k = k.split(", ")
        idx = split_k.index(str(isbn[i]))
        new_keywords.append(" ".join(split_k[idx + 1:]))


    # text 스플릿
    book_sommelier = []
    book_index = []
    book_line = []
    book_review = []
    for i, k in enumerate(text):
        t_1 = p.split(k)
        if len(t_1) == 1:
            book_review.append("")
        elif len(t_1) == 2:
            t_1_s = mo.split(t_1[1])
            if len(t_1_s) == 1:
                book_review.append(t_1[1])
            else:
                book_review.append(d.split(t_1_s[1])[0])

        t_2 = c.split(t_1[0])
        if len(t_2) == 1:
            book_line.append("")
        elif len(t_2) == 2:
            t_2_s = mo.split(t_2[1])
            if len(t_2_s) == 1:
                book_line.append(t_2[1])
            else:
                book_line.append(d.split(t_2_s[1])[0])

        t_3= m.split(t_2[0])
        if len(t_3) == 1:
            book_index.append("")
        elif len(t_3) == 2:
            t_3_s = mo.split(t_3[1])
            if len(t_3_s) == 1:
                book_index.append(t_3[1])
            else:
                book_index.append(d.split(t_3_s[1])[0])

        t_4 = b.split(t_3[0])
        if len(t_4) == 1:
            book_sommelier.append("")
        elif len(t_4) == 2:
            t_4_s = e.split(t_4[1])
            book_sommelier.append(t_4_s[0])

    # 데이터 저장

    data = data[['isbn','keywordpick','keywords']]

    data["keywords"] = new_keywords
    data['index'] = book_index
    data['line'] = book_line
    data['review'] = book_review
    data['sommelier'] = book_sommelier

    data.to_csv("{}_kyobo_final.csv".format(n), index=False)

        
```
